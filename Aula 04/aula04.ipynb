**Descrição do Problema:**

# Classificador

Para facilitar como no exercício anterior vamos criar uma função para execução dos testes.

```python
def testar_mlp(X_train, X_test, y_train, y_test, neuronios_ocultos=[5, 10, 15]):
    """
    Treina e avalia um Perceptron Multicamadas (MLPClassifier) para diferentes quantidades 
    de neurônios na camada oculta e exibe os resultados em um DataFrame e gráficos de perda.

    Parâmetros:
    X_train, X_test - Conjunto de treino e teste para as entradas
    y_train, y_test - Conjunto de treino e teste para as saídas desejadas
    neuronios_ocultos - Lista com os diferentes números de neurônios a serem testados
    """

    resultados = []  # Lista para armazenar os resultados

    plt.figure(figsize=(12, 4))  # Ajusta o tamanho dos gráficos

    for i, n_neurons in enumerate(neuronios_ocultos):
        print(f"\nTreinando MLP com {n_neurons} neurônios na camada oculta...")

        # Configuração do modelo MLP
        mlp = MLPClassifier(
            solver='adam', 
            activation='logistic', 
            alpha=0.1, 
            hidden_layer_sizes=(n_neurons,), 
            batch_size='auto', 
            max_iter=5000, 
            epsilon=1e-6,
            random_state=42
        )

        # Treinamento do modelo
        mlp.fit(X_train, y_train)

        # Predição no conjunto de teste
        y_pred = mlp.predict(X_test)

        # Coleta das métricas
        acuracia = accuracy_score(y_test, y_pred)
        mse = mean_squared_error(y_test, y_pred)
        n_epocas = mlp.n_iter_
        loss_final = mlp.loss_
        func_saida = mlp.out_activation_
        
        # Relatório de classificação
        relatorio = classification_report(y_test, y_pred, target_names=["Tipo A", "Tipo B", "Tipo C"], output_dict=True)

        # Adiciona os resultados na lista
        resultados.append({
            "Neurônios Ocultos": n_neurons,
            "Acurácia": acuracia,
            "MSE": mse,
            "Último Loss": loss_final,
            "Épocas Treinadas": n_epocas,
            "Ativação Saída": func_saida
        })

        # Plotando o gráfico de perda para cada configuração
        plt.subplot(1, len(neuronios_ocultos), i + 1)
        plt.plot(mlp.loss_curve_)
        plt.title(f"Neurônios: {n_neurons}")
        plt.xlabel("Épocas")
        plt.ylabel("Loss")
        plt.grid()

    plt.tight_layout()
    plt.show()

    # Criando o DataFrame com os resultados
    df_resultados = pd.DataFrame(resultados)

    # Exibir os resultados no console
    print("\nResultados do Treinamento:")
    print(df_resultados.to_string(index=False))

    return df_resultados  # Retorna o DataFrame para uso posterior

# Chamando a função (exemplo, caso tenha os dados X_train, X_test, y_train, y_test disponíveis)
# df_resultados = testar_mlp(X_train, X_test, y_train, y_test)
```

Agora vamos executa-la

```python
# Executando o treinamento e avaliação do MLPClassifier com diferentes quantidades de neurônios ocultos
df_resultados = testar_mlp(X_train, X_test, y_train, y_test)

# Exibição do DataFrame com os resultados
print(df_resultados)
```

# Avaliação do modelo

```python
# Identifica a melhor configuração com base na maior acurácia
melhor_config = df_resultados.loc[df_resultados["Acurácia"].idxmax()]

# Obtém o número ideal de neurônios na camada oculta
melhor_n_neurons = int(melhor_config["Neurônios Ocultos"])

print(f"\nMelhor configuração identificada: {melhor_n_neurons} neurônios ocultos.")

# Instancia o modelo MLP com a melhor configuração encontrada
mlp_melhor = MLPClassifier(
    solver='adam', 
    activation='logistic', 
    alpha=0.1, 
    hidden_layer_sizes=(melhor_n_neurons,), 
    batch_size='auto', 
    max_iter=5000, 
    epsilon=1e-6,
    random_state=42
)

# Validação cruzada utilizando todo o dataset com a melhor configuração
print("\nIniciando Validação Cruzada (5-Folds) com a Melhor Configuração...")
scores_cross_val = cross_val_score(mlp_melhor, X, y, cv=5)

# Exibição dos resultados da validação cruzada
print("\nResultados da Validação Cruzada:")
print(f"Scores por fold: {scores_cross_val}")
print(f"Média da Acurácia: {scores_cross_val.mean():.4f}")
print(f"Desvio Padrão: {scores_cross_val.std():.4f}")
```

# Testando o modelo

```python
# Carregar o conjunto de teste externo a partir de um arquivo CSV
# O dataset contém amostras que serão usadas para avaliar o desempenho do modelo treinado
df_teste = pd.read_csv("bebidasTestes.csv")

# Separar as variáveis de entrada (features) e saída (labels) do conjunto de teste externo
# As colunas 'd1', 'd2' e 'd3' representam as classes do conjunto de dados (variáveis de saída)
y_teste_externo = df_teste[['d1', 'd2', 'd3']]  # Rótulos verdadeiros do conjunto de teste

# Removemos as colunas de saída para obter as variáveis de entrada (features) do modelo
X_teste_externo = df_teste.drop(['d1', 'd2', 'd3'], axis=1)  # Atributos usados para predição

# Treinar o modelo MLP com os dados de treino (caso ele ainda não tenha sido treinado)
# Certifique-se de que X_train e y_train foram definidos antes de executar esta etapa
mlp_melhor.fit(X_train, y_train)  # Ajuste do modelo usando o conjunto de treinamento

# Fazer previsões no conjunto de teste externo
# O modelo utilizará X_teste_externo para prever as classes correspondentes
y_pred_externo = mlp_melhor.predict(X_teste_externo)  # Previsões geradas pelo modelo
```

Fazendo as previsões

```python
# ==========================================
# Cálculo das métricas de avaliação do modelo
# ==========================================

# MAE (Mean Absolute Error): Erro absoluto médio, indica o erro médio entre os valores reais e previstos
mae = mean_absolute_error(y_teste_externo, y_pred_externo)

# MSE (Mean Squared Error): Erro médio quadrático, penaliza mais os erros maiores ao elevá-los ao quadrado
mse = mean_squared_error(y_teste_externo, y_pred_externo)

# RMSE (Root Mean Squared Error): Raiz quadrada do MSE, mantém a unidade dos dados originais e dá mais peso a erros grandes
rmse = np.sqrt(mse)

# MAPE (Mean Absolute Percentage Error): Erro absoluto percentual médio, expressa o erro médio em relação aos valores reais
mape = mean_absolute_percentage_error(y_teste_externo, y_pred_externo)

# ==========================================
# Exibição das métricas calculadas
# ==========================================

print("Métricas de Validação (Conjunto Externo):")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.4f}")

# ==========================================
# Visualização da comparação entre valores reais e estimados
# ==========================================

# Criar índices para cada amostra no conjunto de teste externo
indices = np.arange(len(y_teste_externo))  

# Definir a largura das barras para a visualização
largura_barra = 0.4  

# Criar um gráfico de barras para comparar os valores reais e previstos pelo modelo
plt.figure(figsize=(12, 6))  # Define o tamanho do gráfico

# Barras para os valores reais
plt.bar(
    indices - largura_barra / 2,  # Ajusta a posição das barras dos valores reais
    y_teste_externo.sum(axis=1),  # Soma das colunas para representar os valores reais totais
    width=largura_barra,  
    label="Valores Reais",  
    color='blue'  
)

# Barras para os valores previstos
plt.bar(
    indices + largura_barra / 2,  # Ajusta a posição das barras dos valores previstos
    y_pred_externo.sum(axis=1),  # Soma das colunas para representar os valores previstos totais
    width=largura_barra,  
    label="Valores Estimados",  
    color='orange'  
)

# ==========================================
# Configuração e exibição do gráfico
# ==========================================

plt.title("Comparação entre Valores Reais e Estimados (Conjunto Externo)")  # Título do gráfico
plt.xlabel("Índice da Amostra")  # Nome do eixo X
plt.ylabel("Valores")  # Nome do eixo Y
plt.xticks(indices)  # Definir os rótulos do eixo X
plt.legend()  # Adicionar a legenda ao gráfico
plt.grid(True, linestyle="--", alpha=0.5)  # Adicionar uma grade com estilo pontilhado
plt.show()  # Exibir o gráfico
```

```python
# ========================================================
# Cálculo da porcentagem de acertos e erros na predição
# ========================================================

# Número total de amostras no conjunto de teste externo
total_amostras = len(y_teste_externo)

# Contar quantas previsões foram completamente corretas (todas as saídas previstas coincidem com os valores reais)
acertos = (y_teste_externo.values == y_pred_externo).all(axis=1).sum()

# O restante das amostras são consideradas erros (previsões incorretas)
erros = total_amostras - acertos

# Calcular a porcentagem de acertos e erros
percentual_acertos = (acertos / total_amostras) * 100
percentual_erros = (erros / total_amostras) * 100

# ========================================================
# Criação do gráfico de pizza para visualização dos resultados
# ========================================================

plt.figure(figsize=(8, 6))  # Define o tamanho da figura

# Criar gráfico de pizza para representar visualmente os acertos e erros
plt.pie(
    [percentual_acertos, percentual_erros],  # Dados (percentuais)
    labels=["Acertos", "Erros"],  # Rótulos das fatias
    autopct="%1.1f%%",  # Exibir valores percentuais com uma casa decimal
    colors=["green", "red"],  # Cores: verde para acertos, vermelho para erros
    startangle=90,  # Rotaciona o gráfico para que comece do topo
    wedgeprops={"edgecolor": "black"}  # Adiciona borda preta para melhor visibilidade das fatias
)

# ========================================================
# Configuração e exibição do gráfico
# ========================================================

plt.title("Porcentagem de Acertos e Erros da Predição")  # Define o título do gráfico
plt.show()  # Exibe o gráfico
```
