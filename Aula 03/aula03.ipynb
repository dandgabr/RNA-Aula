# Importação das bibliotecas

Carregando os dados de treinamento e de teste que serão utilizados.

```python
# Carregando o conjunto de dados de treinamento
# O arquivo "ressonanciaMLP.csv" contém os dados que serão usados para treinar o modelo
df_treinamento = pd.read_csv("ressonanciaMLP.csv")

# Carregando o conjunto de dados de teste
# O arquivo "ressonanciaMLPTest.csv" contém os dados que serão usados para avaliar o desempenho do modelo treinado
df_teste = pd.read_csv("ressonanciaMLPTest.csv")

# Já que os dados já foram pré-processados as etapas de análise serão suprimidas.
```

# Separação de dados

Separação das colunas de objetivo e dados de entrada.

```python
# Separação dos dados de entrada (features) e saída (target)
# As colunas, exceto "d", são as variáveis independentes (features)
# A coluna "d" é a variável dependente (target) que o modelo deve prever
x = df_treinamento.drop("d", axis=1)  # Remove a coluna "d" para usar as demais como features
y = df_treinamento["d"]  # Define "d" como o target

# Divisão dos dados em treino e teste
# Define 80% dos dados para treino e 20% para teste, com embaralhamento automático
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, random_state=42)

# Verificar o tamanho dos conjuntos
print(f"Tamanho do conjunto de treino: {x_treino.shape[0]} amostras")
print(f"Tamanho do conjunto de teste: {x_teste.shape[0]} amostras")

# Separação dos dados de entrada (features) e saída (target) do conjunto de teste
# A mesma lógica do conjunto de treinamento é aplicada ao conjunto de teste
x_teste_externo = df_teste.drop("d", axis=1)  # Variáveis independentes (features) do conjunto de teste
y_teste_externo = df_teste["d"]  # Variável dependente (target) do conjunto de teste
```

# Função de teste

Gerando função que automatiza os testes

```python
def achar_melhor_configuracao_do_mlp(x_treino, y_treino, x_teste, y_teste):
    """
    Busca a melhor configuração para o MLPRegressor testando diferentes combinações
    de funções de ativação e tamanhos de camada escondida.

    Args:
        x_treino (array-like): Dados de entrada para treinamento.
        y_treino (array-like): Alvos para treinamento.
        x_teste (array-like): Dados de entrada para teste.
        y_teste (array-like): Alvos para teste.

    Returns:
        tuple: Melhor modelo treinado, menor MSE obtido e a melhor configuração usada.
    """

    # Configurações dos parâmetros a serem testados
    funcoes_ativacao = ['tanh', 'relu', 'logistic']  # Tipos de funções de ativação a serem avaliadas
    tamanhos_camadas_escondidas = [5, 10, 15]  # Quantidades de neurônios na camada escondida
    taxa_de_aprendizado = 0.01  # Taxa de aprendizado fixa para todos os testes
    maximo_iteracoes = 1000  # Número máximo de iterações para o treinamento do modelo

    # Variáveis para armazenar os melhores resultados
    melhor_mse = float('inf')  # Inicialmente, um valor muito alto para o erro médio quadrático (MSE)
    melhor_configuracao = None  # Armazena a configuração que resulta no menor MSE
    melhor_modelo = None  # Armazena o melhor modelo treinado

    # Iterar sobre todas as combinações de parâmetros
    for ativacao in funcoes_ativacao:
        for tamanho in tamanhos_camadas_escondidas:
            # Criar o modelo com a combinação atual de parâmetros
            modelo = MLPRegressor(
                activation=ativacao,
                hidden_layer_sizes=(tamanho,),
                learning_rate_init=taxa_de_aprendizado,
                max_iter=maximo_iteracoes,
                random_state=42  # Para garantir reprodutibilidade dos resultados
            )

            # Treinar o modelo com os dados de treinamento
            modelo.fit(x_treino, y_treino)

            # Fazer previsões nos dados de teste
            y_pred = modelo.predict(x_teste)

            # Calcular o erro médio quadrático (MSE) para avaliar o desempenho
            mse = metrics.mean_squared_error(y_teste, y_pred)

            # Verificar se o modelo atual é o melhor encontrado até agora
            if mse < melhor_mse:
                melhor_mse = mse  # Atualizar o menor MSE
                melhor_configuracao = {  # Registrar a configuração que gerou este resultado
                    'Função de ativação': ativacao,
                    'Tamanho da camada escondida': tamanho
                }
                melhor_modelo = modelo  # Salvar o modelo treinado com melhor desempenho

    # Retornar o melhor modelo, o menor MSE e a configuração correspondente
    return melhor_modelo, melhor_mse, melhor_configuracao
```

# Avaliando o melhor modelo

O código abaixo avalia e retorna o melhor modelo



```python
# Chamar a função para buscar a melhor configuração do MLP
melhor_modelo, melhor_mse, melhor_configuracao = achar_melhor_configuracao_do_mlp(x_treino, y_treino, x_teste, y_teste)

# Exibir os resultados
print("Melhor Modelo:", melhor_modelo)
print("Menor MSE:", melhor_mse)
print("Melhor Configuração:", melhor_configuracao)
```
