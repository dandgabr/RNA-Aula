# Importando dependências

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn import metrics
````

# Datasets

````python
# Função para criar dataset com lags (valores defasados)
def create_lagged_df(data, lags=5):
    """
    Gera um DataFrame contendo colunas de valores passados (lags) e a variável alvo.

    Parâmetros:
    data (array-like): Série temporal de entrada.
    lags (int): Número de lags a serem considerados.

    Retorna:
    pd.DataFrame: DataFrame com colunas representando os lags e a variável alvo.
    """
    cols = [f'x-{i}' for i in range(lags, 0, -1)] + ['x']
    lagged_data = [data[i:i+lags+1] for i in range(len(data)-lags)]
    return pd.DataFrame(lagged_data, columns=cols)
````

````python
# Carregar os dados do dataset
df = pd.read_csv('variacaoValores.csv')
````

````python
# Separação dos dados em treino e teste
train, test = df.iloc[:100, 1].values, df.iloc[95:, 1].values
````

````python
# Criando DataFrames com as defasagens
dfNewTrain = create_lagged_df(train, lags=5)
dfNewTest = create_lagged_df(test, lags=5)
````

````python
# Definição das variáveis de entrada e saída
xTrain, yTrain = dfNewTrain.drop('x', axis=1), dfNewTrain['x']
xTest, yTest = dfNewTest.drop('x', axis=1), dfNewTest['x']
````

# Treinamento e resultados

````python
# Definição dos hiperparâmetros para os testes
neuronios = [15, 25, 50]  # Número de neurônios na camada oculta
solvers = ['sgd', 'adam']  # Métodos de otimização a serem testados
resultados = []  # Lista para armazenar os resultados

# Executar treinamentos para cada combinação de neurônios e solver
for n in neuronios:
    for solver in solvers:
        print(f"Treinando MLP com {n} neurônios e solver '{solver}'")

        # Configuração do modelo MLPRegressor
        mlp = MLPRegressor(
            hidden_layer_sizes=(n),
            activation='relu',
            solver=solver,
            max_iter=5000,
            tol=1e-6,
            momentum=0.9,
            early_stopping=True,
            epsilon=1e-6,
            learning_rate_init=0.001,
            learning_rate='constant'
        )

        # Treinamento do modelo
        mlp.fit(xTrain, yTrain)

        # Predição dos valores no conjunto de teste
        y_pred = mlp.predict(xTest)

        # Cálculo das métricas de desempenho
        mae = metrics.mean_absolute_error(yTest, y_pred)
        mse = metrics.mean_squared_error(yTest, y_pred)
        r2 = metrics.r2_score(yTest, y_pred)

        # Armazenamento dos resultados
        resultados.append({
            'Neurônios': n,
            'Solver': solver,
            'Loss': mlp.loss_,
            'Épocas': mlp.n_iter_,
            'MAE': mae,
            'MSE': mse,
            'R²': r2
        })

        # Gráfico de comparação entre valores reais e preditos
        plt.figure(figsize=(10,5))
        plt.bar(range(len(yTest)), yTest, alpha=0.6, label='Real')
        plt.bar(range(len(y_pred)), y_pred, alpha=0.6, label='Predito')
        plt.title(f'Comparação Real vs Predito ({n} neurônios, {solver})')
        plt.legend()
        plt.show()

        # Gráfico da curva de perda durante o treinamento
        plt.figure(figsize=(8,5))
        plt.plot(mlp.loss_curve_, label="Erro durante o treinamento")
        plt.title(f'Curva de Perda ({n} neurônios, {solver})')
        plt.xlabel('Épocas')
        plt.ylabel('Loss')
        plt.legend()
        plt.grid()
        plt.show()
````

````python
# Criar DataFrame com os resultados
df_resultados = pd.DataFrame(resultados)

# Exibir os resultados
print("\nResultados de Treinamento:")
print(df_resultados)
````

# Item 3, 4 e 5

````python
# Definição das topologias das redes neurais
topologias = {
    'Rede 1': 15,
    'Rede 2': 25,
    'Rede 3': 50
}
melhor_solver = 'adam'  # Definição do melhor solver observado

# Lista para armazenar os resultados
resultados = []

# Executar 3 treinamentos para cada rede
for rede, n_neuronios in topologias.items():
    for t in range(1, 4):
        print(f"Treinando {rede} - Execução {t}")

        # Configuração do modelo MLPRegressor com pesos aleatórios
        mlp = MLPRegressor(
            hidden_layer_sizes=(n_neuronios),
            activation='relu',
            solver=melhor_solver,
            max_iter=5000,
            tol=1e-6,
            momentum=0.9,
            early_stopping=True,
            epsilon=1e-6,
            learning_rate_init=0.001,
            learning_rate='constant'
        )

        # Treinamento do modelo
        mlp.fit(xTrain, yTrain)

        # Predição dos valores no conjunto de teste
        y_pred = mlp.predict(xTest)

        # Cálculo das métricas de desempenho
        mae = metrics.mean_absolute_error(yTest, y_pred)
        mse = metrics.mean_squared_error(yTest, y_pred)
        r2 = metrics.r2_score(yTest, y_pred)

        # Registro dos resultados
        resultados.append({
            'Treinamento': f'T{t}',
            'Rede': rede,
            'Loss': mlp.loss_,
            'Épocas': mlp.n_iter_,
            'MAE': mae,
            'MSE': mse,
            'R²': r2,
            'y_pred': y_pred  # Armazenar previsões para gráficos posteriores
        })

# Criar DataFrame com os resultados
df_resultados = pd.DataFrame(resultados)

# Pivotar a tabela para melhor visualização
df_pivot = df_resultados.pivot(index='Treinamento', columns='Rede', values=['Loss', 'Épocas', 'MAE', 'MSE', 'R²'])
print("\nResultados de Treinamento e Avaliação:")
print(df_pivot)
````

````python
# Selecionar o melhor treinamento de cada rede
melhores_treinamentos = df_resultados.loc[df_resultados.groupby('Rede')['Loss'].idxmin()]

# Criar gráficos de barras para comparar os valores reais e preditos
for _, row in melhores_treinamentos.iterrows():
    plt.figure(figsize=(10,5))
    plt.bar(range(len(yTest)), yTest, alpha=0.6, label='Real')
    plt.bar(range(len(row['y_pred'])), row['y_pred'], alpha=0.6, label='Predito')
    plt.title(f'Comparação Real vs Predito ({row["Rede"]}, {row["Treinamento"]})')
    plt.legend()
    plt.show()
````
